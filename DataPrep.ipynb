Ben is messing with the intro text on his own fork.  This notebook links the metadata csv file Speakers_by_session to the speech txt files in the sessions folder. The output is a dataframe with one row per speech, information about the speaker and the text content (uncleaned)

```python
import os, pandas as pd, nltk
```

```python
speechdir = './data/sessions' #path to sessions folder
file = './data/Speakers_by_session.csv' #metadata csv file
```

Useful functions:

```python
def list_files(folder_path):
    """Outputs a list of every file in folder_path or its subdirectories that has a specified extension.
    Prepends specified extension with '.' if it doesn't start with it already.
    If no extension is specified, it just returns all files in folder_path."""
    
    matches = []
    
    for dirpath,dirnames,filenames in os.walk(folder_path):
        for filename in filenames:
            matches.append(os.path.join(dirpath,filename))
    
    return matches
```

```python
def load_txt(file_path):
    """Loads text (string) into memory. Must be assigned to object."""
    
    with open(file_path, encoding="utf8") as file_handler:
        text = file_handler.read()
    return text
```

Creating a dataframe and matching each speaker to the content of their speech:

```python
df = pd.read_csv(file) #opens the csv file in dataframe format
df['file_end']=df['ISO Code'].astype(str) +"_"+ df['Session'].apply(lambda x: "{0:.0f}".format(x)).astype(str) +"_"+ df['Year'].apply(lambda x: "{0:.0f}".format(x)).astype(str)+'.txt'
#adds a column that corresponds to ISO Code_Session_Year to be matched with txt files

filelist=list_files(speechdir)
file_end=[filename[-15:] for filename in filelist] #ends of file name = ISO Code_Session_Year
file_text=[load_txt(filename) for filename in filelist] #loads the content of each file
df_text=pd.DataFrame({'file_end':file_end, 'text':file_text}) #list of file content in dataframe

df_ungdc = pd.merge(df, df_text, on='file_end', how='inner', indicator=False) #merge with metadate frame on ISO COde_Session_Year identifier
df_ungdc
```
